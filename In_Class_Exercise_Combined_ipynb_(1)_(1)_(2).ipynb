```python
# Write your answer here
# Import libraries
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer

# Download VADER lexicon
nltk.download('vader_lexicon')

# Initialize VADER
sia = SentimentIntensityAnalyzer()

# Create new column 'compound'
df['compound'] = df['Text'].apply(lambda x: sia.polarity_scores(x)['compound'])

# Show first 5 rows of Text and compound
print(df[['Text', 'compound']].head())
```

### Q3. Create a sentiment label from the compound score (3 points)
Create a new column called `PredictedSentiment` using this rule:
- `compound >= 0.05` → `positive`
- `compound <= -0.05` → `negative`
- otherwise → `neutral`

Then show the first 10 rows of `Text`, `compound`, and `PredictedSentiment`.

```python
# Write your answer here
# Function to convert compound score into sentiment label
def get_sentiment_label(score):
    if score >= 0.05:
        return "positive"
    elif score <= -0.05:
        return "negative"
    else:
        return "neutral"

# Create new column
df['PredictedSentiment'] = df['compound'].apply(get_sentiment_label)

# Show first 10 rows
print(df[['Text', 'compound', 'PredictedSentiment']].head(10))
```

### Q4. Compare review score and sentiment (2 points)
Group by `Score` and compute the **average compound score** for each score value.

```python
# Write your answer here
# Group by Score and compute average compound sentiment
avg_sentiment_by_score = df.groupby('Score')['compound'].mean()

# Display result
print(avg_sentiment_by_score)
```

Your prompt should include:
- a clear instruction

- a request for a clean output format

Store your prompt in a Python variable called `zero_shot_prompt`, then print it.

```python
# Write your answer here
zero_shot_prompt = """
Classify each product review as one of the following labels: positive, negative, or neutral.

Instructions:
- Read the review carefully.
- Assign exactly one label: positive, negative, or neutral.
- Do not include explanations.

Output Format:
Return the result in this format only:
Review: <text>
Sentiment: <label>

Reviews:
1. The product works perfectly and I am very happy with it.
2. The quality is bad and it stopped working after one week.
3. It is okay, not great but not terrible either.
"""

print(zero_shot_prompt)
```

## Reflection (optional, no extra points)
Briefly explain the difference between:
- sentiment analysis
- text classification
- prompt engineering

Sentiment analysis is a specific type of text analysis that focuses on identifying the emotional tone of a text, such as positive, negative, or neutral. Text classification is a broader task where text is assigned to any predefined category, such as spam vs not spam, topic labels, or sentiment labels. Prompt engineering is the process of designing and structuring inputs (prompts) to guide an AI model to produce better, more accurate, and more consistent outputs for tasks like sentiment analysis or text classification.
