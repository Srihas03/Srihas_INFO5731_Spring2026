Using RegEx, extract and create two new columns:

```python
# Setup (run this cell first)
import re
import pandas as pd
```

## Q1 (1 point) — Answer below

```python
# Q1 — ANSWER CELL
FILE_PATH = "/support_messages.txt.txt"


# Load TAB-separated file
df = pd.read_csv(FILE_PATH, sep="\t", names=["id", "message"])

# Print outputs
print(df.shape)
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
# Q2 — ANSWER CELL

# word_count: number of words
df["word_count"] = df["message"].apply(lambda x: len(str(x).split()))

# char_count: number of characters
df["char_count"] = df["message"].apply(lambda x: len(str(x)))

# num_count: number of digits (0–9)
df["num_count"] = df["message"].apply(lambda x: sum(c.isdigit() for c in str(x)))

# upper_word_count: number of ALL-CAPS words
df["upper_word_count"] = df["message"].apply(
    lambda x: sum(word.isupper() for word in str(x).split())
)

# Print full DataFrame
print(df)
```

## Q3 (3 points) — Answer below

```python
# Q3 — ANSWER CELL

import re
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS

STOPWORDS = set(ENGLISH_STOP_WORDS)

def clean_text(text: str) -> str:
    text = str(text)

    # 1. lowercase
    text = text.lower()

    # 2. remove punctuation/symbols (keep letters/numbers/spaces)
    text = re.sub(r'[^a-z0-9\s]', '', text)

    # 3. remove English stopwords
    words = text.split()
    words = [word for word in words if word not in STOPWORDS]

    # 4. remove extra spaces
    text = " ".join(words)

    return text

# Create clean column
df["clean"] = df["message"].apply(clean_text)

# Print original and clean for id=1 and id=4
print(df[df["id"] == 1][["message", "clean"]])
print(df[df["id"] == 4][["message", "clean"]])
```

## Q4 (2 points) — Answer below

```python
# Q4 — ANSWER CELL

import re

# order_id: first occurrence of ORD-#### (case-insensitive)
df["order_id"] = df["message"].str.extract(
    r"(ORD-\d{4})",
    flags=re.IGNORECASE
)

# email: first email address if present
df["email"] = df["message"].str.extract(
    r"([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})"
)

# Display results
print(df[["id", "order_id", "email"]])
```

## Q5 (1 point) — Answer below

```python
# Q5 — ANSWER CELL

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# 1) Fit TF-IDF on df["clean"]
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df["clean"])

# 2) Compute average TF-IDF per term across documents
feature_names = vectorizer.get_feature_names_out()
avg_tfidf = tfidf_matrix.mean(axis=0).A1   # average per column

# 3) Print top 5 terms + their average scores
top_indices = np.argsort(avg_tfidf)[-5:][::-1]

print("Top 5 terms by average TF-IDF:")
for idx in top_indices:
    print(feature_names[idx], round(avg_tfidf[idx], 4))
```
